# Pairgrid

```python
iris.shape
```

```python
x = sns.PairGrid(iris)
x = x.map(plt.scatter)
```

```python
#Mapeia na diagonal e fora da diagonal
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
#Add hue
x = sns.PairGrid(iris, hue='species', palette='husl')
x = x.map_diag(plt.hist, histtype='step', linewidth=2.8)
x = x.map_offdiag(plt.scatter)
x = x.add_legend() #Tem que adicionar
```

```python
# Filtra variaveis
x = sns.PairGrid(iris, vars=['petal_length', 
                             'petal_width'],
                               hue='species')
x = x.map_diag(plt.hist, edgecolor='blue') #edgecolor altera a cor das linhas das barras
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
x = sns.PairGrid(iris, x_vars=['petal_length', 
                               'petal_width'], 
                 y_vars=['sepal_length',
                         'sepal_width'])
x = x.map(plt.scatter)
```

```python
# Em cima e abaixo da diagonal diferentes
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

```python
# hue_kws = muda os marcadores para cada quebra
x = sns.PairGrid(iris, hue='species', 
                 palette='RdBu',
                 hue_kws={'marker':['d','s','+']}) #s = square
x = x.map_diag(plt.hist, edgecolor='black')
x = x.map_upper(plt.scatter, s=100)#tamanho do marcador
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
diamonds_df = sns.load_dataset('diamonds')
diamonds_df.describe(include=['category'])
```

```python
cuts = diamonds_df.cut.unique()
clarities = diamonds_df.clarity.unique()
for cut, clarity in zip(cuts, clarities):
    plt.figure(figsize=(4,4))
    mask = ((diamonds_df.cut==cut) & (diamonds_df.clarity==clarity))
    filtered = diamonds_df.loc[mask]
    sns.boxplot(data=filtered, 
                x='depth',
                palette='pastel')
    plt.title(cut+'_'+clarity)
    plt.show()
```
